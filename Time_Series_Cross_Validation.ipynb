```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit
```

```python
# Generate fake time series data
data = np.random.rand(100)
```

# Time Series Split Cross-Validation

```python
tscv = TimeSeriesSplit(n_splits=5, gap=-1)

# Create a list of colors for train and test data
colors = ['blue', 'green', 'red', 'orange', 'purple']

# Plot the data
fig, ax = plt.subplots(nrows=5, figsize=(8, 10))  # Adjust the figure size if needed
for i, (train_index, test_index) in enumerate(tscv.split(data)):
    X_train, X_test = data[train_index], data[test_index]
    ax[i].plot(train_index, X_train, color=colors[i], label=f'Train {i+1}')
    ax[i].plot(test_index, X_test, color=colors[i], linestyle='dashed', label=f'Test {i+1}')

    # Add title and labels
    ax[i].set_title(f'Fold {i+1} Data')
    ax[i].set_ylabel('Value')

# Add a legend
plt.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

# Blocked Cross-Validation
set `max_train_size` to balance training folds

```python
# Split the data using TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5, gap=-1, max_train_size=30)

# Create a list of colors for train and test data
colors = ['blue', 'green', 'red', 'orange', 'purple']

# Plot the data
fig, ax = plt.subplots(nrows=5, figsize=(8, 10))  # Adjust the figure size if needed
for i, (train_index, test_index) in enumerate(tscv.split(data)):
    X_train, X_test = data[train_index], data[test_index]
    ax[i].plot(train_index, X_train, color=colors[i], label=f'Train {i+1}')
    ax[i].plot(test_index, X_test, color=colors[i], linestyle='dashed', label=f'Test {i+1}')

    # Add title and labels
    ax[i].set_title(f'Fold {i+1} Data')
    ax[i].set_ylabel('Value')

# Add a legend
plt.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```
